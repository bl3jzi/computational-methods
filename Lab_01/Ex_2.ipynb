## Subpoint A:

### Libraries used:

```python
from decimal import Decimal, getcontext
from math import sqrt
```

**Setting decimal precision:**

```python
getcontext().prec = 16
```

```python
x = Decimal("124819e-14")
```

**Equations:**

```python
original = (x+1).sqrt() - 1
stable = x / ((x+1).sqrt() + 1)
```

**Answers:**

```python
print("Original equation: ", original)
print("Stable equation:   ", stable)
```
